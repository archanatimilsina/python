```python
import cv2

# 1. Load the pre-trained "Face Brain" (XML file)
# The shortcut 'cv2.data.haarcascades' automatically finds where the file is on your PC
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# 2. Load your image
# Make sure "aprina.jpg" is in the same folder as this script!
img = cv2.imread('aprina.jpg')

# 3. Convert to Grayscale
# Face detection works better and faster on black-and-white images
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 4. Detect Faces
# scaleFactor: 1.1 means it looks for faces of different sizes
# minNeighbors: 5 means it checks 5 times before deciding it's definitely a face
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

# 5. Draw Rectangles around the faces
for (x, y, w, h) in faces:
    # (255, 0, 0) is the color BLUE in BGR format
    # 2 is the thickness of the line
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)

# 6. Show the result
cv2.imshow('Face Detection Preview', img)
cv2.waitKey(0) # Press any key to close the window
cv2.destroyAllWindows()
```
