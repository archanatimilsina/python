```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
import pandas as pd
import numpy as np
np.set_printoptions(suppress=True, precision=4)
```

```python
# load dataset
datasetArray = load_breast_cancer()
# print(datasetArray.data)
dataset = pd.DataFrame(datasetArray.data, columns=datasetArray.feature_names)
# print(dataset.head())
print(dataset.columns)
print(dataset.shape)
# x= dataset.data
# y= dataset.target
x= dataset
y= datasetArray.target
```

```python
# split the data and do scaling
xtr, xte, ytr, yte = train_test_split(x,y, test_size=0.1, random_state=42)
scaler= StandardScaler()
xtr_scaled= scaler.fit_transform(xtr)
xte_scaled = scaler.transform(xte)

```

```python
# train the model
model = LogisticRegression()
model.fit(xtr_scaled,ytr)
prediction = model.predict(xte_scaled)
```

```python
# Evaluation
print(f"Accuracy Score:{accuracy_score(yte, prediction):.2%}")
```
