```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
data ={
    "Age": [19, 21, 20, 23, 31, 22, 35, 29, 30, 40],
    "AnnualIncome":[15, 16, 17, 18, 45, 23, 55, 48, 50, 60],
    "City":["Kathmandu", "Pokhara", "Lalitpur", "Kathmandu", "Pokhara", 
             "Lalitpur", "Kathmandu", "Pokhara", "Kathmandu", "Lalitpur"],
      "SpendingScore":[39, 81, 6, 77, 40, 76, 6, 94, 3, 72]
}
df= pd.DataFrame(data)
print(df)
numerical_features = ['Age','AnnualIncome','SpendingScore']
catagorical_features = ['City']
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("num",StandardScaler(),numerical_features),
        ("cat",OneHotEncoder(drop="first"),catagorical_features)
    ]
)
X=preprocessor.fit_transform(df)
```

```python
model = KMeans(n_clusters=3,
               init="k-means++",
               n_init=10,
               random_state=42,
               max_iter=300
               )
labels = model.fit_predict(X)
print(labels)
```
