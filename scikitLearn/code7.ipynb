```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import seaborn as svn
import pandas as pd
from sklearn.svm import SVC
```

```python
# load dataset
df = svn.load_dataset("tips")
# df.columns
y=df['smoker']

```

```python
# df['total_bill'].unique() # no need
# df['tip'].unique() # no need
# df['sex'].unique() # need ['Male', 'Female']
# df['smoker'].unique() #  need ['No', 'Yes']
# df['day'].unique() #  need ['Sun', 'Sat', 'Thur', 'Fri']
# df['time'].unique() #  need ['Dinner', 'Lunch']
# df['size'].unique() # no need [2, 3, 4, 1, 6, 5]
```

```python
en_need = ['sex','day','time']
ohe= OneHotEncoder(sparse_output=False, drop='first')
x_encoded =ohe.fit_transform(df[en_need])
x_encoded_df = pd.DataFrame(x_encoded , columns=ohe.get_feature_names_out(en_need))
x_final = pd.concat([df.drop(en_need,axis=1).reset_index(drop=True),x_encoded_df],axis=1)
X = x_final.drop('smoker', axis=1) # You must remove the target from the features
y = x_final['smoker']
```

```python
xtr,xte,ytr,yte = train_test_split(X,y,test_size=0.2,random_state=33)
scaler = StandardScaler()
xtr_scaled = scaler.fit_transform(xtr)
xte_scaled = scaler.transform(xte)
```

```python
kernels =['linear','poly','rbf']
for k in kernels:
    model = SVC(kernel= k)
    model.fit(xtr_scaled,ytr)
    prediction = model.predict(xte_scaled)
    accuracy = accuracy_score(yte, prediction)
    print(f"Accuracy with kernel {k} is {accuracy:.2%}")
```
