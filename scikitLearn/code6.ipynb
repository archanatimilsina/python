```python
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# load titanic dataset from seaborn
dataset1 = sns.load_dataset('titanic')
print(dataset1.isna().sum())
dataset =dataset1.drop(['deck','embark_town'],axis=1)
```

```python
print(dataset.isna().sum())
```

```python
dataset['age']=dataset['age'].fillna(dataset['age'].median())
dataset['embarked']=dataset['embarked'].fillna(dataset['embarked'].mode()[0])
# dataset = dataset.dropna(subset=dataset['embarked'])

```

```python
le= LabelEncoder()
dataset['sex']= le.fit_transform(dataset['sex'])
dataset['embarked']= le.fit_transform(dataset['embarked'])
dataset['who']= le.fit_transform(dataset['who'])
dataset['class']= le.fit_transform(dataset['class'])
dataset['adult_male']= le.fit_transform(dataset['adult_male'])
dataset['alive']= le.fit_transform(dataset['alive'])
dataset['alone']= le.fit_transform(dataset['alone'])
# print(dataset)
```

```python
X = dataset.drop(['survived', 'alive', 'who', 'adult_male'], axis=1)
y= dataset['survived']
# print(X)
```

```python
xtr, xte, ytr, yte= train_test_split(X,y , random_state=33, test_size=0.2)
```

```python
# scale splitted data 
scaler = StandardScaler()
xtr_scaled = scaler.fit_transform(xtr)
xte_scaled = scaler.transform(xte)
```

```python
model1 = RandomForestClassifier(n_estimators=100, random_state=33)
model2 = SVC(kernel='rbf',C=1.0)
model3 = MLPClassifier(
    hidden_layer_sizes=(100,50),
    max_iter=1000,
    random_state=33
)

models = {
    'Random Forest': model1,
    'SVM' : model2,
    'multi-layer perceptron':model3
}

for name,model in models.items():
    model.fit(xtr_scaled,ytr)
    prediction = model.predict(xte_scaled)
    accuracy = accuracy_score(yte, prediction)
    print(f"Accuracy Score for {name} model is :{accuracy}")
```
