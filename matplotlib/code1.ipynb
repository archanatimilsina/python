```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10,4))

# subplot(nrows, ncols, index)
plt.subplot(1, 2, 1)  # 1 row, 2 cols, first subplot
plt.plot(x, y1, color='blue')
plt.title('Sine Function')

plt.subplot(1, 2, 2)  # 1 row, 2 cols, second subplot
plt.plot(x, y2, color='orange')
plt.title('Cosine Function')

plt.tight_layout()  # avoids overlap
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10,8))  # 2x2 grid

axes[0,0].plot(x, np.sin(x))
axes[0,0].set_title('sin(x)')

axes[0,1].plot(x, np.cos(x), color='orange')
axes[0,1].set_title('cos(x)')

axes[1,0].plot(x, np.sin(2*x), color='green')
axes[1,0].set_title('sin(2x)')

axes[1,1].plot(x, np.cos(2*x), color='red')
axes[1,1].set_title('cos(2x)')

fig.tight_layout()
plt.show()
```

```python
functions = [np.sin, np.cos, lambda x: np.sin(2*x), lambda x: np.cos(2*x)]
titles = ['sin(x)','cos(x)','sin(2x)','cos(2x)']

fig, axes = plt.subplots(2,2, figsize=(10,8))

for ax, func, title in zip(axes.flat, functions, titles):
    ax.plot(x, func(x))
    ax.set_title(title)
    ax.grid(True)

fig.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
y_positive = np.array([1, 2, 3, 4, 5])
y_negative = -np.array([1, 2, 1, 3, 2])
x = np.arange(5)

plt.fill_between(x, 0, y_positive, color='green', alpha=0.5, label='Positive')
plt.fill_between(x, 0, y_negative, color='red', alpha=0.5, label='Negative')
plt.title("Positive and Negative Area Plot")
plt.legend()
plt.show()
```

```python
days = np.arange(1, 6)
apples = np.array([3, 4, 2, 5, 6])
oranges = np.array([1, 2, 3, 2, 1])
bananas = np.array([2, 1, 4, 1, 2])

plt.stackplot(days, apples, oranges, bananas, labels=['Apples','Oranges','Bananas'], colors=['#FF9999','#FFA500','#FFFF99'], alpha=0.7)
plt.xlabel("Days")
plt.ylabel("Quantity")
plt.title("Stacked Area Plot of Fruits")
plt.legend(loc='upper left')
plt.show()

data = np.array([apples, oranges, bananas])
total = data.sum(axis=0)
percent_data = data / total  # convert to fractions

plt.stackplot(days, percent_data[0], percent_data[1], percent_data[2],
              labels=['Apples','Oranges','Bananas'],
              colors=['#FF9999','#FFA500','#FFFF99'], alpha=0.7)
plt.ylabel("Percentage")
plt.xlabel("Days")
plt.title("Percentage Stacked Area Plot")
plt.legend(loc='upper left')
plt.show()

plt.stackplot(days, apples, oranges, bananas, labels=['Apples','Oranges','Bananas'],
              colors=['#FF9999','#FFA500','#FFFF99'],
              alpha=0.7, edgecolor='black', linewidth=0.5)
plt.xlabel("Days")
plt.ylabel("Quantity")
plt.title("Styled Stacked Area Plot")
plt.legend(loc='upper left')
plt.show()

plt.stackplot(days, apples, oranges, bananas, labels=['Apples','Oranges','Bananas'], colors=['#FF9999','#FFA500','#FFFF99'], alpha=0.5)
plt.plot(days, apples, 'r-o')
plt.plot(days, oranges, 'g-s')
plt.plot(days, bananas, 'b-^')
plt.title("Stacked Area with Line Overlays")
plt.legend(loc='upper left')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, color='blue', label='sin(x)')
plt.fill_between(x, y, color='blue', alpha=0.3)  # fill area under curve
plt.title("Basic Area Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, color='blue', label='3D Line')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title("3D Line Plot")
ax.legend()
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
colors = np.random.rand(50)
sizes = np.random.randint(20, 200, size=50)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(x, y, z, c=colors, s=sizes, cmap='viridis', alpha=0.8)
fig.colorbar(sc, label='Random Value')
ax.set_title("3D Scatter Plot")
plt.show()
```

```python
X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.8)
fig.colorbar(surf, shrink=0.5, aspect=10)
ax.set_title("3D Surface Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

fig = plt.figure(figsize=(10,6))
gs = gridspec.GridSpec(2, 2, figure=fig)  # 2 rows, 2 columns

ax1 = fig.add_subplot(gs[0, :])   # top row spans both columns
ax2 = fig.add_subplot(gs[1, 0])   # bottom-left
ax3 = fig.add_subplot(gs[1, 1])   # bottom-right

ax1.plot(x, y1, color='blue')
ax1.set_title("Top Plot (Spans Two Columns)")

ax2.plot(x, y2, color='green')
ax2.set_title("Bottom-left Plot")

ax3.plot(x, y3, color='red')
ax3.set_title("Bottom-right Plot")

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8,6), sharex=True, sharey=True)

for i in range(2):
    for j in range(2):
        axs[i,j].plot(x, np.sin(x + i + j))
        axs[i,j].set_title(f'Plot {i},{j}')

plt.tight_layout()
plt.show()
```
