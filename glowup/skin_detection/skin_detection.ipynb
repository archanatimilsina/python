```python
import cv2
import numpy as np
```

```python
def read_and_resize_image(image_path, max_width = 400, max_height = 500):
    img = cv2.imread(image_path , cv2.IMREAD_COLOR)
    if img is None:
        raise FileNotFoundError(f"Could not find or open image at: {image_path}")
    height, width = img.shape[:2]
    print(width, height)
    f1 = max_width/width
    f2 = max_height/height
    f= min(f1,f2)
    new_dim = int(width*f), int(height*f)
    print(new_dim)
    resized_img = cv2.resize(img, new_dim, interpolation=cv2.INTER_AREA)
    return resized_img
    
```

```python
img =read_and_resize_image("./aprina.jpg")
print(img)
```

```python
img = cv2.imread("./aprina.jpg" , cv2.IMREAD_COLOR)
img_grayscale= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
histogram, bin_edges = np.histogram(img_grayscale,256,[0, 256])
```
