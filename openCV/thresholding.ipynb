```python
import cv2
import numpy as np
```

```python
img = cv2.imread("./aprina.jpg" , cv2.IMREAD_COLOR)
img_grayscale= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# print(img_grayscale[156])
```

```python
# Binary Thresholding
ret_threshold,thresholded_image = cv2.threshold(img_grayscale, 200 , 255, cv2.THRESH_BINARY)
print(thresholded_image[156])
print(ret_threshold)
```

```python
# Inverse binary thresholding
ret_threshold , thresholded_image = cv2.threshold(img_grayscale, 200, 255, cv2.THRESH_BINARY_INV)
print(thresholded_image[156])
print(ret_threshold)
```

```python
# Otsu thresholding
ret_threshold , thresholded_image = cv2.threshold(img_grayscale, 200, 255, cv2.THRESH_OTSU)
print(thresholded_image[156])
print(ret_threshold)
```

```python
# Otsu thresholding + Binary inverse
ret_threshold , thresholded_image = cv2.threshold(img_grayscale, 200, 255, cv2.THRESH_OTSU+ cv2.THRESH_BINARY_INV)
print(thresholded_image[156])
print(ret_threshold)
```
