```python
import cv2
import matplotlib.pyplot as plt
```

```python
# 1. Load image in grayscale
image = cv2.imread('./assets/aprina.png', cv2.IMREAD_GRAYSCALE)  # replace with your image path
```

```python
threshold_value = 128
_, binary_image= cv2.threshold(image, threshold_value,212, cv2.THRESH_BINARY)
```

```python
# cv2.imshow("Binary Image",binary_image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Original GrayScale")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title(f"Thresholded at {threshold_value}")
plt.imshow(binary_image, cmap='gray')
plt.axis('off')

plt.show()
```
